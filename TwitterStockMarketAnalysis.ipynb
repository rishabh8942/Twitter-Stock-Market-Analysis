***Task 2: Statistical Tests and Analysis***

**1.Give me the Z-test or T-test over High, Low, and
Close columns and see if the null hypothesis gets
rejected or accepted:**


```python
from scipy import stats

# Perform T-test between 'High' and 'Low'
t_stat, p_value = stats.ttest_ind(df_cleaned['High'], df_cleaned['Low'])
print(f"T-test between High and Low: t-stat = {t_stat:.4f}, p-value = {p_value:.4f}")

# Perform T-test between 'High' and 'Close'
t_stat, p_value = stats.ttest_ind(df_cleaned['High'], df_cleaned['Close'])
print(f"T-test between High and Close: t-stat = {t_stat:.4f}, p-value = {p_value:.4f}")

# Perform T-test between 'Low' and 'Close'
t_stat, p_value = stats.ttest_ind(df_cleaned['Low'], df_cleaned['Close'])
print(f"T-test between Low and Close: t-stat = {t_stat:.4f}, p-value = {p_value:.4f}")
```

**2. Check if the data is dependent or independent by
using the chi-square method:**

```python
from scipy.stats import chi2_contingency

# Convert numerical columns into categorical bins
df_cleaned['High_Category'] = pd.qcut(df_cleaned['High'], q=3, labels=["Low", "Medium", "High"])
df_cleaned['Low_Category'] = pd.qcut(df_cleaned['Low'], q=3, labels=["Low", "Medium", "High"])
df_cleaned['Close_Category'] = pd.qcut(df_cleaned['Close'], q=3, labels=["Low", "Medium", "High"])

# Create a contingency table (e.g., between High and Close categories)
contingency_table = pd.crosstab(df_cleaned['High_Category'], df_cleaned['Close_Category'])

# Perform Chi-Square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Print results
print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"P-Value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies Table:\n", expected)

# Interpretation of results
if p_value < 0.05:
    print("\nConclusion: Reject the null hypothesis  The data is dependent.")
else:
    print("\nConclusion: Fail to reject the null hypothesis  The data is independent.")
```

**3. Assign buttons to control time periods. Add the
buttons to analyze the stock prices of Twitter in
different time periods:**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display

# Convert 'Date' to datetime format explicitly
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

# Function to filter data based on selected time period
def filter_data(period):
    if period == "Last 6 Months":
        return df[df['Date'] >= df['Date'].max() - pd.DateOffset(months=6)]
    elif period == "Last 1 Year":
        return df[df['Date'] >= df['Date'].max() - pd.DateOffset(years=1)]
    elif period == "Last 3 Years":
        return df[df['Date'] >= df['Date'].max() - pd.DateOffset(years=3)]
    else:
        return df  # Full Data

# Function to update the plot dynamically
def update_plot(period):
    filtered_df = filter_data(period)

    plt.figure(figsize=(12, 6))
    sns.lineplot(x=filtered_df['Date'], y=filtered_df['Close'], label="Closing Price", color="blue")

    # Formatting
    plt.xlabel("Year")
    plt.ylabel("Closing Price (USD)")
    plt.title(f"Twitter Stock Closing Prices Over Time ({period})")
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)

    plt.show()

# Create interactive buttons for different time periods
buttons = widgets.ToggleButtons(
    options=["Last 6 Months", "Last 1 Year", "Last 3 Years", "Full Data"],
    description="Select Period:",
    style={'description_width': 'initial'}
)

# Display the interactive buttons and update plot dynamically
display(buttons)
widgets.interactive(update_plot, period=buttons)
```
